```python
import numpy as np
import matplotlib.pyplot as plt

from genedata import generater
from nonparam import KernalSmooth
from semiparam import WeightFunction

import warnings
warnings.filterwarnings("ignore")
```

```python
n = 300
p = 10
k = 6
rou = 0.5
alpha = 0.7

y,X,U  = generater(n,p,k,rou)
```

```python
wf = WeightFunction()
wf.fit(X,U,y,0.3,alpha)

linear_hat = np.dot(X,wf.beta)
g_hat = KernalSmooth().fitpredict(U, y-linear_hat, U, 0.3, "gaussian")

plt.scatter(U,y)
plt.scatter(U,linear_hat + g_hat)
plt.show()
```
